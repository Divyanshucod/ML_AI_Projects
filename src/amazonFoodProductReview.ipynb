```python
## libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Preprocessing and cleaning
CommentsReviews = df[['Id','Text','Score']]
```

```python
CommentsReviews.isnull().sum()
```

```python
def Conversion(score):
    if score <= 2:
        return 0
    elif score <=3:
        return 1
    else:
        return 2
```

```python
## feature engineering
CommentsReviews['SentimentScore'] = CommentsReviews['Score'].apply(Conversion)
```

```python
CommentsReviews
```

```python
CommentsReviews.drop(columns=['Score'],axis=1,inplace=True)
```

```python
CommentsReviews.to_csv('validatedReviews.csv',index=False)
```

```python
CommentsReviews = CommentsReviews.head(100000)
```

```python
## independent and dependent features
X = CommentsReviews[['Text']]
y = CommentsReviews[['SentimentScore']]
```

```python
## train test split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,y,test_size=0.20)
```

```python
X_train
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
wordnetLemma = WordNetLemmatizer()
```

```python
import re
```

```python
## convert to cleaned text
def cleaned_text(text):
    text = text.lower()
    text = re.sub(r"[^a-z\s]", "", text)
    words = text.split()
    stop_words = set(stopwords.words("english"))
    words = [w for w in words if not w in stop_words]
    words = [wordnetLemma.lemmatize(word,'v') for word in words]
    return ' '.join(words)
```

```python
## for x_train 
Reviews_X_train = []
for review in X_train['Text']:
    Reviews_X_train.append(cleaned_text(review))
```

```python
## for x_test
Reviews_X_test = []
for review in X_test['Text']:
    Reviews_X_test.append(cleaned_text(review))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=5000) 
```

```python
from gensim.models import Word2Vec 
```

```python
X_train['Text'].tolist()
```

```python
## Select Vectorizing method
## TF-IDF
def TFIdf(x_train,x_test):
     X_train_tfidf = tfidf_vectorizer.fit_transform(x_train) 
     X_test_tfidf = tfidf_vectorizer.transform(x_test)
     return X_train_tfidf, X_test_tfidf

def Word2vec(x_train):
    # Create a list of sentences 
    train_sentences = x_train['Text'].tolist()

    # Train Word2Vec model
    model = Word2Vec(sentences=train_sentences, vector_size=200, window=5, min_count=1, workers=4) 

    return model
    
```

```python
## vectorization (converting words to vectors)
train_tfidf,test_tfidf = TFIdf(Reviews_X_train,Reviews_X_test)
```

```python
## model training
from sklearn.linear_model import LogisticRegression
regressionLog = LogisticRegression()
```

```python
train_tfidf = train_tfidf.toarray()
test_tfidf = test_tfidf.toarray()
```

```python
Y_train.shape
```

```python
regressionLog.fit(train_tfidf,Y_train)
```

```python
y_pred = regressionLog.predict(test_tfidf)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print(confusion_matrix(Y_test,y_pred))
print(accuracy_score(Y_test,y_pred))
```
